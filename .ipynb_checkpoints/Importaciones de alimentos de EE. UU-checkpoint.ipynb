```python
# Crear una lista para almacenar las longitudes de los textos
lengths = []

# Iterar sobre cada fila del DataFrame
for index, row in df_Titles.iterrows():
    # Verificar si el valor en la columna 'description' es un string
    if isinstance(row['description'], str):
        # Si es un string, calcular su longitud y añadirla a la lista
        text_length = len(row['description'])
        lengths.append(text_length)
    else:
        # Si no es un string (por ejemplo, es un NaN o un float), añadir un valor predeterminado a la lista
        lengths.append(0)  # Puedes elegir otro valor predeterminado si lo prefieres

# Imprimir las longitudes de los textos
print("Longitudes de los textos en la columna 'description':")
print(lengths)
```

```python

# Convertir la lista 'lengths' en una serie de Pandas
lengths_series = pd.Series(lengths)
```
